```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
#Scaler import
#standard - Uses Mean / Standard deviation
#MinMax - Uses min/max
#Robust - Uses Quartile

std = StandardScaler()
```

```python
rb = RobustScaler()
rb_scaled = rb.fit_transform(df)
rb_scaled = pd.DataFrame(rb_scaled, columns = df.columns)
rb_scaled
```

```python
round(rb_scaled.describe().T, 2)
```

```python
mm = MinMaxScaler()
mm_scaled = mm.fit_transform(df)
mm_scaled = pd.DataFrame(mm_scaled, columns = df.columns)
mm_scaled
```

```python
from sklearn.model_selection import train_test_split

xtr,xt,ytr,yt = train_test_split(df.drop('class', axis = 1), df['class'], test_size = 0.2, random_state = 100)
```

```python
mm.fit(xtr)
```

```python
xtr_scaled = mm.transform(xtr)
xt_scaled = mm.transform(xt)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 7)
knn.fit(xtr_scaled, ytr)
pred = knn.predict(xt_scaled)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(pred, yt)
```

```python
# This code evaluates the performance of a K-Nearest Neighbors (KNN) classifier for different values of `n_neighbors` (from 1 to 20).
# For each value of `n_neighbors`:
# 1. A KNN model is trained on the scaled training data (`xtr_scaled`, `ytr`).
# 2. The model is used to predict labels for the scaled test data (`xt_scaled`).
# 3. The accuracy of the predictions is calculated using `accuracy_score` and stored in the `scores` list.
# Finally, the list `scores` contains the accuracy scores for each value of `n_neighbors`, which can be used to determine the optimal number of neighbors.

scores = []
for i in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors = i)
    knn.fit(xtr_scaled, ytr)
    pred = knn.predict(xt_scaled)
    acc = accuracy_score(pred, yt)
    scores.append(acc)

scores
```

```python
#Visualization
sns.lineplot(x = range(1,21), y = scores)
plt.show()
```

# Since no further improvement is observed after n_neighbors = 13, it has been confirmed that 13 is a reasonable hyperparameter.

```python
#remodeling

knn = KNeighborsClassifier(n_neighbors = 13)
knn.fit(xtr_scaled, ytr)
pred = knn.predict(xt_scaled)
accuracy_score(yt, pred)
```
