```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from google.colab import drive

```

```python
data = pd.read_csv('/content/sample_data/Salary_Data.csv')
data.head()
```

```python
print(data.isnull().sum())
```

```python
figure = px.scatter(data_frame = data,
                    x="Salary",
                    y="YearsExperience",
                    size="YearsExperience",
                    trendline="ols")
figure.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = np.asanyarray(data[["YearsExperience"]])
y = np.asanyarray(data[["Salary"]])
xtrain, xtest, ytrain, ytest = train_test_split(x, y,
                                                test_size=0.2,
                                                random_state=42)
```

```python
model = LinearRegression()
model.fit(xtrain, ytrain)
```

```python
a = float(input("Years of Experience : "))
features = np.array([[a]])
print("Predicted Salary = ", model.predict(features))
```
